## Assignment

Import necessary library

```python
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
```

Read the data

```python
df_can = pd.read_excel('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx',
                       sheet_name='Canada by Citizenship',
                       skiprows=range(20),
                       skipfooter=2)
```

Drop unnecessary columns

```python
# in pandas axis=0 represents rows (default) and axis=1 represents columns.
df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
df_can.head(2)
```

Rename columns title

```python
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df_can.columns
```

Add a 'Total' column

```python
df_can['Total'] = df_can.sum(axis=1)
```

**Question 1:** Let's compare the number of immigrants from India and China from 1980 to 2013.


Step 1: Get the data set for China and India, and display dataframe.

```python
### type your answer here
china_data = df_can[df_can['Country'] == 'China'].transpose()
india_data = df_can[df_can['Country'] == 'India'].transpose()

china = china_data.drop(china_data.index[0:4])
china = china.drop(["Total"])

india = india_data.drop(china_data.index[0:4])
india = india.drop(["Total"])

china.rename(columns={'36': 'noOfImmigrants'}, inplace=True)
india.rename(columns={'79': 'noOfImmigrants'}, inplace=True)

print("China Data:")
print(china_data)

print("\nIndia Data:")
print(india_data)
```

```python
print(df_can.columns)
years = [1980, 1981, 1982, 1983, 1984, 1985, 1986, 1987, 1988, 1989, 1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013]
#df_can.index.values
df_can.set_index('Country', inplace=True)
#df_can.index.values
```

Step 2: Plot graph. We will explicitly specify line plot by passing in `kind` parameter to `plot()`.

```python
### type your answer here
import matplotlib.pyplot as plt

# Plotting data for China
china.plot(kind='line')

# Customize the plot
plt.title('Immigration from China')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')
plt.legend().remove()
plt.show()

# Plotting data for India
india.plot(kind='line')

# Customize the plot
plt.title('Immigration from India')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')
plt.legend().remove()
plt.show()
```

**Question 2:** Compare the trend of top 5 countries that contributed the most to immigration to Canada.

Step 1: Get the data set for top 5 countries

```python
df_can.sort_values(['Total'], ascending=False, axis=0, inplace=True)

# get the top 5 entries
df_top5 = df_can.head()

# transpose the dataframe
df_top5 = df_top5[years].transpose()

df_top5
```

Step 2: Plot graph

```python
### type your answer here
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='area',
             stacked=False,
             figsize=(20, 10),
             )

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

**Question 3**: Create an unstacked area plot of the 5 countries that contributed the least to immigration to Canada **from** 1980 to 2013. Use a transparency value of 0.55.

```python
### type your answer here
df_can.sort_values(['Total'], ascending=True, axis=0, inplace=True)
df_least5 =  df_can.head()

df_least5 = df_least5[years].transpose()
#print(df_least5)

df_least5.plot(kind='area', alpha=0.55, stacked=False, figsize=(10, 6))

plt.title('Immigration Trend of 5 Countries with Least Contribution to Canada')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')
plt.show()

```

**Question 4**: Display the immigration distribution for Greece, Albania, and Bulgaria for years 1980 - 2013? Use an overlapping plot with 15 bins and a transparency value of 0.35.

```python
### type your answer here
countries = ['Greece', 'Albania', 'Bulgaria']
df_countries = df_can.loc[countries, years]
df_countries = df_countries.transpose()
df_countries.plot(kind='hist', bins=15, alpha=0.35, figsize=(10, 6))
plt.title('Immigration Distribution for Greece, Albania, and Bulgaria (1980 - 2013)')
plt.xlabel('Number of Immigrants')
plt.ylabel('Frequency')
plt.show()
```

**Question 5:** Create a *horizontal* bar plot showing the *total* number of immigrants to Canada from the top 15 countries, for the period 1980 - 2013. Label each country with the total immigrant count.

Step 1: Get the data pertaining to the top 15 countries.

```python
### type your answer here
df_top15 = df_can.sort_values(by='Total', ascending=False).head(15)
df_top15
```

Step 2: Plot data:
   1. Use `kind='barh'` to generate a bar chart with horizontal bars.
   2. Make sure to choose a good size for the plot and to label your axes and to give the plot a title.
<!--    3. Loop through the countries and annotate the immigrant population using the anotate function of the scripting interface. -->

```python
### type your answer here
df_top15['Total'].plot(kind='barh', figsize=(15, 9), color='skyblue')
df_top15 = df_top15.sort_values(by='Total', ascending=True)
plt.xlabel('Number of Immigrants')
plt.ylabel('Country')
plt.title('Top 15 Countries by Total Immigrants to Canada (1980 - 2013)')
for i, val in enumerate(df_top15['Total']):
    plt.text(val, i, f'{val:,}', fontsize=10, va='center')
plt.show()
```
