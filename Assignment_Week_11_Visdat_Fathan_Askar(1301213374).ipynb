

* Buatlah visualisasi jaringan yang merepresentasikan jalur transportasi yang menghubungkan kota-kota besar di Jawa Barat.
* Tentukan kota terpenting berdasarkan nilai degree centrality dan betweenness centrality.
* Update ukuran node berdasarkan nilai degree centrality



```python
#Gambar untuk refrensi jalur transportasi (Jalur utama Jawa Barat dan Jalur Kereta Jawa Barat)
import cv2
from google.colab.patches import cv2_imshow
import networkx as nx
import matplotlib.pyplot as plt

jalanTol=cv2.imread("/content/JalurJabar.jpg")
jalurKereta=cv2.imread("/content/JalurjabarKereta.jpg")

new_size = (800, 600)
jalanTol = cv2.resize(jalanTol, new_size)
jalurKereta = cv2.resize(jalurKereta, new_size)
cv2_imshow(jalanTol)
cv2_imshow(jalurKereta)
```

```python
#Jalur utama Jawa Barat(garis warna merah)

# Membuat grafik kosong
G = nx.Graph()

# Menambahkan node (kota)
cities = ["Bandung","Banjar", "Bogor", "Bekasi", "Depok", "Cirebon", "Sukabumi", "Tasikmalaya", "Ciamis",
          "Purwakarta", "Karawang", "Sumedang", "Indramayu"]

G.add_nodes_from(cities)

# Menambahkan edge (jalur transportasi)
connections = [("Bogor", "Depok"), ("Bogor", "Cianjur"), ("Bogor", "Sukabumi"), ("Bekasi", "Karawang"),
               ("Karawang", "Purwakarta"), ("Karawang", "Indramayu"), ("Karawang", "Cirebon"),
               ("Sukabumi", "Cianjur"), ("Cianjur", "Bandung"), ("Cianjur", "Purwakarta"), ("Purwakarta", "Indramayu"),
               ("Purwakarta", "Bandung"), ("Purwakarta", "Cirebon"), ("Bandung", "Sumedang"),
               ("Bandung", "Tasikmalaya"), ("Bandung", "Ciamis"), ("Sumedang", "Cirebon"),
               ("Tasikmalaya", "Ciamis"), ("Ciamis", "Banjar")]

G.add_edges_from(connections)

# Menggambar grafik
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, font_size=10)
plt.show()
```

```python
# Degree Centrality
degree_centrality = nx.degree_centrality(G)
print(degree_centrality)
# Betweenness Centrality
betweenness_centrality = nx.betweenness_centrality(G)
print(betweenness_centrality)
# Menentukan kota terpenting
most_important_city_degree = max(degree_centrality, key=degree_centrality.get)
most_important_city_betweenness = max(betweenness_centrality, key=betweenness_centrality.get)

print(f"Kota terpenting berdasarkan degree centrality: {most_important_city_degree}")
print(f"Kota terpenting berdasarkan betweenness centrality: {most_important_city_betweenness}")
```

```python
# Menghitung ukuran node berdasarkan degree centrality
node_size = [v * 3000 for v in degree_centrality.values()]

# Menggambar grafik dengan ukuran node yang diupdate
plt.figure(figsize=(12, 8))
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray', node_size=node_size, font_size=10)
plt.title('Jalur utama Jawa Barat dengan Ukuran Node Berdasarkan Degree Centrality')
plt.show()
```

```python
#Kereta Jawa Barat
G = nx.Graph()
nodes = {
    'Bogor': 2, 'Depok': 2, 'Bandung': 4, 'Purwakarta': 1, 'Cianjur': 1,
    'Sukabumi': 1, 'Cimahi': 2, 'Cikampek': 1, 'Karawang': 1,
    'Arjawinangun': 1, 'Cirebon': 1, 'Tasikmalaya': 1, 'Ciamis': 1, 'Banjar': 1
}

for node, size in nodes.items():
    G.add_node(node, size=size * 300)

edges = [
    ('Depok', 'Bogor'), ('Bogor', 'Sukabumi'), ('Sukabumi', 'Cianjur'),
    ('Cianjur', 'Cimahi'), ('Cimahi', 'Purwakarta'), ('Cimahi', 'Bandung'),
    ('Purwakarta', 'Cikampek'),('Cikampek', 'Karawang'),('Karawang', 'Bekasi'),
    ("Cikampek","Arjawinangun"),("Arjawinangun","Cirebon"),("Bandung","Tasikmalaya"),
    ("Tasikmalaya", "Ciamis"),("Ciamis","Banjar")
]

G.add_edges_from(edges)
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, font_size=10)
plt.show()
```

```python
# Degree Centrality
degree_centrality = nx.degree_centrality(G)
print(degree_centrality)
# Betweenness Centrality
betweenness_centrality = nx.betweenness_centrality(G)
print(betweenness_centrality)
# Paling penting
most_important_city_degree = max(degree_centrality, key=degree_centrality.get)
most_important_city_betweenness = max(betweenness_centrality, key=betweenness_centrality.get)

print(f"Kota terpenting berdasarkan degree centrality: {most_important_city_degree}")
print(f"Kota terpenting berdasarkan betweenness centrality: {most_important_city_betweenness}")
```

```python
node_size = [v * 3000 for v in degree_centrality.values()]

plt.figure(figsize=(12, 8))
nx.draw(G, with_labels=True, node_size=node_size, node_color='skyblue', edge_color='gray')
plt.title('Jaringan Kereta Api Jawa Barat dengan Ukuran Node Berdasarkan Degree Centrality')
plt.show()
```
